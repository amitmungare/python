### Loops

```python
## for loop
for i in range(5):
    print(i)
```

```python

for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(10, 1, -1):
    print(i)
```

```python
str="Hello, World!"
for i in str:
    print(i)
```

```python
## While loop
count = 0
while count < 5:
    print(count)
    count += 1
    
```

```python
count = 0
while count%2 == 0:
    print(count)
    count += 1
```

```python
## loop control statements
## break statement
for i in range(10):
    if  i==5:
        break
    print(i)
    
```

```python
## continue statement
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
## pass: used as a placeholder for future code. It does nothing when executed.
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loops
for i in range(3):
    for j in range(2):
        print(f"i:{i}, j:{j}")
```

```python
## Example: sum of first n natural numbers using for loop
n=10
sum=0
for i in range(1, n+1):
    sum +=i

print(f"sum of first {n} natural numbers is {sum}") 
```

```python
## Example: prime numbers between 1 and 100 using for loop
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```
