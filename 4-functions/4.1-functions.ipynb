# Functions

#### function is a reusable block of organized code

```python
## syntax

def function_name(name):
    return f'your name is {name}'

function_name("amit")
```

```python
## multi parameters
def sum(num1, num2):
    return num1+num2
```

```python
total = sum(2,4)
print(total)
```

```python
## default parameters

def greet(name="Guest"):
    return f"Hello {name}"
```

```python
print(greet())
print(greet("amit"))
```

```python
## variable length arguments

def print_num(*args):
    for num in args:
        print(num, end=" ")
```

```python
print_num(1,2,4,6,7,8,9, 'amit', "sam")
```

```python
## keywords arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f'{key}- {value}')
```

```python
print_details(name="amit", age=26, city="Goa")
```

```python
## return multipal parameters

def ret(a, b):
    return a*b, a, b
```

```python
print(ret(4,10))
```
