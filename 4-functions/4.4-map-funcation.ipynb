### Map() function

```python
## syntax

# map(function, list)
```

```python
nums=[1,2,3,4,5,6]

def sqr(num):
    return num*num

my_map = map(sqr, nums)
print(list(my_map))
```

```python
## lambda function with map

sqr1 = lambda num: num*num
my_map1= map(sqr1, nums)
print(list(my_map1))

print(list(map(lambda x:x*x, nums)))    
```

```python
## map multiple iterables

num1=[1,2,3,4]
num2=[5,6,7,8]
# print(list(map(lambda args:exp, args )))
print(list(map(lambda n1,n2:n1+n2, num1, num2 )))
```

```python
## map() to convert the list of string to interger

str1=['1', '2', '3', '4']

print(list(map(int, str1)))
```
