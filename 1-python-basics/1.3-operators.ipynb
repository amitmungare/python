### Operators

```python
## Arithethic operators
a=10
b=5
add_result=a+b # addition
sub_result=a-b # subtraction
mul_result=a*b # multiplication
div_result=a/b # division
floor_div_result=a//b  # floor division
mod_result=a%b # modulus
exp_result=a**b # exponentiation

print("Addition:", add_result)
print("Subtraction:", sub_result)
print("Multiplication:", mul_result)
print("Division:", div_result)
print("Floor Division:", floor_div_result)
print("Modulus:", mod_result)
print("Exponentiation:", exp_result)
```

```python
## Comparison operators
x=10
y=20
print("x > y:", x > y)  # Greater than
print("x < y:", x < y)  # Less than
print("x == y:", x == y)  # Equal to
print("x != y:", x != y)  # Not equal to
print("x >= y:", x >= y)  # Greater than or equal to
print("x <= y:", x <= y)  # Less than or equal to
print("x is y:", x is y)  # Identity operator
print("x is not y:", x is not y)  # Identity operator
print("x in [10, 20, 30]:", x in [10, 20, 30])  # Membership operator
print("y not in [10, 20, 30]:", y not in [10, 20, 30])  # Membership operator

```

### Logical Operator

```python
x=True
y=False
print("x and y:", x and y)  # Logical AND
print("x or y:", x or y)  # Logical OR
print("not x:", not x)  # Logical NOT
```
