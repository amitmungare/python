```python
## type inference

var1 = 10
print(var1, type(var1))
var2 = "Hello"
print(var2, type(var2))
var3 = 3.14
print(var3, type(var3))
var4 = True
print(var4, type(var4))
var5 = [1, 2, 3]
print(var5, type(var5))
var6 = (1, 2, 3)
print(var6, type(var6))
var7 = {"name": "Amit", "age": 25}
print(var7, type(var7))
var8 = {1, 2, 3}
print(var8, type(var8))
var9 = None
print(var9, type(var9))
var10 = lambda x: x * 2
print(var10, type(var10))
var11 = (x for x in range(5))
print(var11, type(var11))
var12 = b"Hello"
print(var12, type(var12))
var13 = bytearray(b"Hello")
print(var13, type(var13))
var14 = memoryview(b"Hello")
print(var14, type(var14))
var15 = frozenset([1, 2, 3])
print(var15, type(var15))
var16 = complex(1, 2)
print(var16, type(var16))
var17 = range(10)
print(var17, type(var17))
var18 = slice(1, 10, 2)
print(var18, type(var18))
var19 = type(var1)
print(var19, type(var19))
var20 = type(var2)
print(var20, type(var20))
```
