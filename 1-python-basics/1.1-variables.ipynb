```python
## Declaration and assignment of variables
age = 25
name = "Amit Mungare"
height = 5.9
is_student = True

print("age:", age)
print("name:", name)
print("height:", height)
print("is_student:", is_student)
```

```python
## Naming conventions
# variable names should be descriptive
# they must start with a letter or an underscore
# they can contain letters, digits, and underscores
# they are case-sensitive
first_name = "Amit"
last_name = "Mungare"
full_name = first_name + " " + last_name
print("Full Name:", full_name)

# valid variable names
x = 10
y = 20
z = x + y
print("Sum:", z)

# invalid variable names
# 1name = "Amit" # starts with a digit
# my-name = "Amit" # contains a hyphen
# my name = "Amit" # contains a space
# @ge = 25 # starts with a special character

## case sensitivity
var = 10
Var = 20
print("var:", var)
print("Var:", Var)
```

```python
## understanding variable types
## Python is a dynamically typed language, which means you don't need to declare the type of a variable. The type is determined at runtime based on the value assigned to it.
x = 10 # x is an integer
print("x:", x, "type:", type(x))
```

```python
## Type checking and type conversion
a = 10
b = 5.5
print("a:", a, "type:", type(a))
print("b:", b, "type:", type(b))    

# Type conversion
a_float = float(a) # convert integer to float
b_int = int(b) # convert float to integer
print("a_float:", a_float, "type:", type(a_float))
print("b_int:", b_int, "type:", type(b_int))

age="25"
print("age:", age, "type:", type(age))
age_int = int(age)
print("age_int:", age_int, "type:", type(age_int))
```

```python
## Dynamic Typing
## Python allows the type of a variable to be changed at runtime
x = 10
print("x:", x, "type:", type(x))
x = "Hello, World!"
print("x:", x, "type:", type(x))
```

```python
## Simple calculator
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

sum = num1+num2
diff = num1-num2
product = num1*num2
quotient = num1/num2

print("Sum:", sum)
print("Difference:", diff)
print("Product:", product)
print("Quotient:", quotient)
```
