### Dictionaries

```python
## creating Dictionaries

empty_dict=dict()
print(type(empty_dict))
emp_dict={}
print(type(emp_dict))
```

```python
student={"name":"amit", "age":26, "grade":99.4}
print(student)
print(type(student))
```

```python
student1={"name":"amit", "age":26, "name":99.4}
print(student1)
```

```python
## accessing dict ele
student={"name":"amit", "age":26, "grade":'A'}
print(student['name'])

print(student.get('age'))
print(student.get('last_name', "NA"))
```

```python
## modifing dicitonary element

print(student)
student['age']=28 # update
student['address']='India' #added
del student['grade'] #delete

print(student)
```

```python
## Dict methods

keys=student.keys()
print(keys)

values=student.values()
print(values)

items=student.items()
print(items)
```

```python
## shallow copy

student_copy = student
print(student)
print(student_copy)

student['name']="krish"
print(student)
print(student_copy)

student_copy1 = student.copy()
student['name'] = "sam"
print(student)
print(student_copy)
print(student_copy1)


```

```python
## Iterating 
for key,value in student.items():
    print(f'{key}- {value}')

print("-----------------------")
for key in student.keys():
    print(key)

print("-----------------------")
for value in student.values():
    print(value)
```

```python
## nested disctionaries

students={
    'student1':{
        "name":"amit", "age":24
    },
    'student2':{
        "name":"sam", "age":26
    },
    'student3':{
        "name":"tom", "age":45
    },
    'student4':{
        "name":"muh", "age":36
    },
}

print(students)
print(students["student2"]['name'])
```

```python
## Iterating nested disctionaries
for id,info in students.items():
    print("---------------")
    for name,age in info.items():
        print(f'{name}- {age}')
```

```python
## dict comphrehension

squares={x:x**2 for x in range(5) if x%2==0}
print(squares)
```

```python
## Merge 2 dict
dict1={'a':1, 'b':2}
dict2={'b':3, 'c':4}
merged_dict={**dict1, **dict2}
print(merged_dict)
```

```python
##
numbers=[1,2,2,3,3,3,4,4,4,4]
freq={}

for num in numbers:
    if num in freq:
        freq[num]+=1
    else:
        freq[num]=1

print(freq)
```
