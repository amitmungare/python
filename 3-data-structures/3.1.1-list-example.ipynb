### Example: Manage a ToDo List

```python
## create a to do list to keep track of tasks
tasks = []

## add tasks to the list
tasks.append("Buy groceries")
tasks.append("Clean the house")
tasks.append("Finish homework")
tasks.append("Call mom")
tasks.append("Pay bills")
tasks.append("Go for a run")

## remove a task from the list
tasks.remove("Clean the house")

## check if a task is in the list
if "Buy groceries" in tasks:
    print("Don't forget to buy groceries!")

## print all tasks
print("To Do List:")
for task in tasks:
    print(f"- {task}")

```

```python
## Organize student grades

grades = [85, 92, 78, 90, 88, 95, 80]

## calculate average grade
average_grade = sum(grades) / len(grades)
print(f"Average Grade: {average_grade:.2f}")

## filter out grades below 90
high_grades = [grade for grade in grades if grade >= 90]
print("High Grades (90 and above):", high_grades)

## higest grade
higest_grade = max(grades)
print(higest_grade)

## lowest grade
lowest_grade = min(grades)
print(lowest_grade)

```

```python
# Managing an inventory
inventory = ["apples", "bananas", "oranges", "grapes"]

# Adding a new item
inventory.append("strawberries")

# Removing an item that is out of stock
inventory.remove("bananas")

# Checking if an item is in stock
item = "oranges"
if item in inventory:
    print(f"{item} are in stock.")
else:
    print(f"{item} are out of stock.")

# Printing the inventory
print("Inventory List:")
for item in inventory:
    print(f"- {item}")

```

```python
# Collecting user feedback
feedback = ["Great service!", "Very satisfied", "Could be better", "Excellent experience"]

# Adding new feedback
feedback.append("Not happy with the service")

# Counting specific feedback
positive_feedback_count = sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())
print(f"Positive Feedback Count: {positive_feedback_count}")

# Printing all feedback
print("User Feedback:")
for comment in feedback:
    print(f"- {comment}")
```
