# Tuple

```python
## creating a tuple

tpl=()
print(tpl)
print(type(tpl))
```

```python
my_tup = tuple([1,2,3,4,5])
print(my_tup)
```

```python
mixed_tup = (1, "hello", 2.3)
print(mixed_tup)
```

```python
tup=(1,2,3,4,5,6,7,8)
print(tup[2])
print(tup[-1])
```

```python
## tuple oprations

concatenation = mixed_tup+tup
print(concatenation)
```

```python
mixed_tup*3
```

```python
## immutable nature of tuples
tup[2]='amit'
```

```python
## tuple methods

print(tup.count(1))
print(tup.index(4))
```

```python
## packing and unpacking tuple

packing_tuple = 1, 'hello', 2.4 #packing
print(packing_tuple)

a,b,c = packing_tuple #unpacking
print(a)
print(b)
print(c)
```

```python
# unpacking with *
numbers =(1,2,3,4,5,6)
first,*middle,last = numbers
print(first)
print(middle)
print(last)
```

```python
##nesting tuple

tups = ((1,2,3),(4,5,6),(4,5,6,7))
print(tups[1])
print(tups[2][0:2])
```

```python
## iterating over tuples

for sub_tup in tups:
    for tup in sub_tup:
        print(tup, end=" ")
    print()
```
