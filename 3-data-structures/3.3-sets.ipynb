### Set


```python
my_set={1,2,3,4}
print(my_set)
print(type(my_set))
```

```python
my_empty_set = set()
print(my_empty_set)
```

```python
my_set={1,2,34,2,43,5,43}
print(my_set)

my_set=set([1,2,34,2,43,5,43])
print(my_set)
```

```python
## basics set opration
my_set.add(7) # add 
print(my_set)

my_set.remove(34) #remove
print(my_set)

my_set.discard(345) #discard



```

```python
## pop element
removed_ele = my_set.pop()
print(removed_ele)
print(my_set)
```

```python
## clear all elements

my_set.clear()
print(my_set)
```

```python
## set memebership

my_set={1,2,3,4,5}
print(3 in my_set)
print(8 in my_set)
```

```python
## Mathematical Operation
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

## union
union_set=set1.union(set2)
print(union_set)

## Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

## Difference
diff_set1 = set1.difference(set2)
print(diff_set1)
diff_set2 = set2.difference(set1)
print(diff_set2)

## symmetric difference
symm_diff = set1.symmetric_difference(set2)
print(symm_diff)
```

```python
## set methods

x={1,2,3}
y={3,4,5}

## is subset
print(set1.issubset(set2))

## is superset
print(set1.issuperset(set2))
```

```python
## remove duplicate elements

lst=[1,2,4,3,4,5,6,76,3,5,5,2,4]
print(set(lst))
```

```python
### Counting unique words in text

text="there are unique words in this text if there are then good"
words = text.split()

unique_words = set(words)
print(unique_words)
print(len(unique_words))
```
